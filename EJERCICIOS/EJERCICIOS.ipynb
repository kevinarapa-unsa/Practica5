```python
#1.programa que ingrese por teclado un número inicial y un número final para generar un rango 

# Función para verificar si un número es par
def es_par(numero):
    return numero % 2 == 0

# Función para verificar si un número es impar
def es_impar(numero):
    return numero % 2 != 0

# Función principal para generar el rango y recorrerlo
def generar_rango():
    # Solicitar los números al usuario
    numero_inicial = int(input("Ingrese el número inicial: "))
    numero_final = int(input("Ingrese el número final: "))
    
    # Generar el rango de números
    numero_actual = numero_inicial
    
    # Recorrer el rango utilizando un ciclo while
    while numero_actual <= numero_final:
        # Verificar si el número es par o impar y mostrar el resultado
        if es_par(numero_actual):
            print(f"{numero_actual} es par.")
        else:
            print(f"{numero_actual} es impar.")
        
        # Incrementar el número actual para continuar el ciclo
        numero_actual += 1

# Llamar a la función para ejecutar el programa
generar_rango()
```

```python
# Función para imprimir la tabla de multiplicar de un número

def tabla_multiplicar(numero):
    # Iniciar el contador en 1
    contador = 1
    
    while contador <= 10:
        resultado = numero * contador
        print(f"{numero} x {contador} = {resultado}")
        contador += 1

def obtener_numero():
    # Solicitar el número al usuario
    numero = int(input("Ingrese un número para ver su tabla de multiplicar: "))
    tabla_multiplicar(numero)

# Llamar a la función principal para ejecutar el programa
obtener_numero()


```

```python
# Función para verificar si una cadena contiene solo letras y espacios

def es_valido(nombre):
    # Comprobamos si todos los caracteres son letras o espacios
    return nombre.replace(" ", "").isalpha()

# Función principal para solicitar el nombre y apellido
def obtener_nombre_apellido():
    # Bucle para solicitar y verificar el nombre
    while True:
        nombre = input("Ingrese su nombre: ")
        if es_valido(nombre):
            break
        else:
            print("El nombre no es válido. Asegúrese de que solo contenga letras y espacios.")
    
    # Bucle para solicitar y verificar el apellido
    while True:
        apellido = input("Ingrese su apellido: ")
        if es_valido(apellido):
            break
        else:
            print("El apellido no es válido. Asegúrese de que solo contenga letras y espacios.")
    
    # Mostrar el nombre y apellido ingresados
    print(f"Nombre completo ingresado: {nombre} {apellido}")

# Llamar a la función para ejecutar el programa
obtener_nombre_apellido()
```

```python
# Función para verificar si el comentario contiene palabras negativas

def tiene_palabras_negativas(comentario):
    palabras_negativas = ["malo", "pesimo", "no", "horrible", "pobre", "terrible", "deficiente"]
    comentario_lower = comentario.lower()  # Convertir el comentario a minúsculas para hacer la búsqueda insensible al caso
    for palabra in palabras_negativas:
        if palabra in comentario_lower:
            return True
    return False

# Función para verificar si el comentario contiene palabras positivas
def tiene_palabras_positivas(comentario):
    palabras_positivas = ["bueno", "excelente", "bien", "genial", "maravilloso", "positivo", "fantástico"]
    comentario_lower = comentario.lower()  # Convertir el comentario a minúsculas para hacer la búsqueda insensible al caso
    for palabra in palabras_positivas:
        if palabra in comentario_lower:
            return True
    return False

# Función principal para solicitar el comentario y clasificarlo
def verificar_comentario():
    while True:
        comentario = input("Ingrese el comentario del alumno EPIT: ")
        
        if tiene_palabras_negativas(comentario):
            print("Comentario NEGATIVO")
            break
        elif tiene_palabras_positivas(comentario):
            print("Comentario POSITIVO")
            break
        else:
            print("Comentario NEUTRO, no contiene palabras positivas ni negativas. Intente nuevamente.")
            # El ciclo continuará pidiendo un comentario si no se detecta ninguna palabra positiva o negativa

# Llamar a la función para ejecutar el programa
verificar_comentario()
```

```python
#lista aleatoria hasta 60 números

import random

# Función para generar una lista de números aleatorios entre 1 y 60
def generar_lista_aleatoria():
    lista = [random.randint(1, 60) for _ in range(60)]  # Genera hasta 60 números aleatorios entre 1 y 60
    return lista

# Función para encontrar múltiplos de 2
def lista_multiplos_2(lista):
    multiplos_2 = []
    i = 0
    while i < len(lista):
        if lista[i] % 2 == 0:
            multiplos_2.append(lista[i])
        i += 1
    return multiplos_2

# Función para encontrar múltiplos de 3
def lista_multiplos_3(lista):
    multiplos_3 = []
    i = 0
    while i < len(lista):
        if lista[i] % 3 == 0:
            multiplos_3.append(lista[i])
        i += 1
    return multiplos_3

# Función para encontrar múltiplos de 5
def lista_multiplos_5(lista):
    multiplos_5 = []
    i = 0
    while i < len(lista):
        if lista[i] % 5 == 0:
            multiplos_5.append(lista[i])
        i += 1
    return multiplos_5

# Función principal para ejecutar el programa
def ejecutar_programa():
    # Generar la lista aleatoria
    lista_aleatoria = generar_lista_aleatoria()
    
    # Dividir la lista en las tres categorías
    multiplos_2 = lista_multiplos_2(lista_aleatoria)
    multiplos_3 = lista_multiplos_3(lista_aleatoria)
    multiplos_5 = lista_multiplos_5(lista_aleatoria)
    
    # Imprimir los resultados
    print("Lista completa de números aleatorios:")
    print(lista_aleatoria)
    
    print("\nMúltiplos de 2:")
    print(multiplos_2)
    
    print("\nMúltiplos de 3:")
    print(multiplos_3)
    
    print("\nMúltiplos de 5:")
    print(multiplos_5)

# Llamar a la función principal para ejecutar el programa
ejecutar_programa()
```

```python
# Función para verificar si un número es capicúa
def es_capicua(numero):
    # Convertir el número a cadena para compararlo
    numero_str = str(numero)
    return numero_str == numero_str[::-1]  # Compara si el número es igual a su reverso

# Función principal para solicitar los números y recorrer el rango
def verificar_capicuas():
    # Solicitar al usuario el número inicial y final
    numero_inicial = int(input("Ingrese el número inicial (3 cifras): "))
    numero_final = int(input("Ingrese el número final (4 cifras): "))
    
    # Verificar que los números estén dentro del rango válido
    if numero_inicial < 100 or numero_inicial > 999:
        print("El número inicial debe ser de 3 cifras.")
        return
    if numero_final < 1000 or numero_final > 9999:
        print("El número final debe ser de 4 cifras.")
        return
    
    # Generar el rango de números
    numero_actual = numero_inicial
    
    # Recorrer el rango utilizando un ciclo while
    while numero_actual <= numero_final:
        # Verificar si el número es capicúa
        if es_capicua(numero_actual):
            print(f"{numero_actual} es un número CAPICÚA.")
        else:
            print(f"{numero_actual} no es un número CAPICÚA.")
        
        # Incrementar el número actual
        numero_actual += 1

# Llamar a la función principal para ejecutar el programa
verificar_capicuas()
```

```python
# Función para verificar si el DNI contiene solo números y tiene exactamente 8 dígitos
def es_dni_valido(dni):
    # Verificar si el DNI tiene exactamente 8 caracteres y si todos son dígitos
    return dni.isdigit() and len(dni) == 8

# Función principal para solicitar y verificar el DNI
def obtener_dni():
    while True:
        # Solicitar el ingreso del DNI
        dni = input("Ingrese su DNI (8 dígitos numéricos): ")
        
        # Verificar si el DNI es válido
        if es_dni_valido(dni):
            print("DNI válido ingresado.")
            break
        else:
            print("DNI inválido. Asegúrese de ingresar un DNI de 8 dígitos numéricos.")
            # El ciclo continuará solicitando el DNI hasta que sea válido

# Llamar a la función principal para ejecutar el programa
obtener_dni()
```
